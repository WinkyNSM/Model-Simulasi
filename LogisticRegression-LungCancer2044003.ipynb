```python
import pandas as pd
df = pd.read_csv("survey lung cancer.csv")
```

```python
df.head(100)
```

```python
df.info()
df["GENDER"].replace(["M","F"], [0,1], inplace=True)
df["LUNG_CANCER"].replace(["YES","NO"], [1,0], inplace=True)
```

```python
df = pd.get_dummies(df)
df
```

```python
prediction = ['GENDER','AGE','SMOKING','YELLOW_FINGERS','CHRONIC DISEASE','WHEEZING','COUGHING','SHORTNESS OF BREATH','SWALLOWING DIFFICULTY','CHEST PAIN']
x = df[prediction]

y = df['LUNG_CANCER']

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x)
scaled_df = scaler.transform(x)
scaled_df = pd.DataFrame(scaled_df, columns= x.columns)
scaled_df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(scaled_df, y , test_size=0.9, random_state=42)
print(len(y_test))
```

```python
from sklearn import linear_model

model = linear_model.LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
