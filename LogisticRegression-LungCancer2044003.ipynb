```python
import pandas as pd
df = pd.read_csv("survey lung cancer.csv")
```

```python
df.columns = ["GENDER","AGE","SMOKING","YELLOW_FINGERS","ANXIETY","PEER_PRESSURE","CHRONIC DISEASE","FATIGUE","ALLERGY","WHEEZING","ALCOHOL CONSUMING","COUGHING","SHORTNESS OF BREATH","SWALLOWING DIFFICULTY","CHEST PAIN","LUNG_CANCER"]
df.info()
df.head()
df["GENDER"].replace(["M","F"], [0,1], inplace=True)
```

```python
data = df.columns
data = pd.get_dummies(data)
data
```

```python
prediction = ["GENDER","AGE","SMOKING","YELLOW_FINGERS","CHRONIC DISEASE","ALLERGY","WHEEZING","COUGHING","SHORTNESS OF BREATH","SWALLOWING DIFFICULTY","CHEST PAIN"]
x = data[prediction]

y = data['LUNG_CANCER']

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x)
scaled_data = scaler.transform(x)
scaled_data = pd.DataFrame(scaled_data, columns= x.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(scaled_data, y , test_size=0.2, random_state=1)
print(len(y_test))
```

```python
from sklearn import linear_model

model = linear_model.LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
